2.1) Un potencial cumple la ecuación de Laplace si su gradiente al cuadrado es igual a 0.

al derivar el potencial dos veces en x se obtiene: A cosh(k(h+z)) (-cos(kh)k^2) cos(wt)

al derivar el potencial dos veces en z se obtiene: A cosh(k(h+z))k^2 cos(kh) cos(wt)

por lo tanto al sumar ambas expreciones se obtiene 0, demostrando que satisface la ecuación de Laplace. Por lo tanto, es un potencial de velocidad de un flujo irrotacional permanente de un fludo incompresible.


2.2) En el fondo de la piscina z = -h.

potencial = A(-1)cos(kx)cos(wt)


2.3) En la pared izquierda x = 0 y x = L en la derecha. y su velocidad en x debe ser 0.
potencial_i = Acosh(k(h+k))cos(wt)
potencial_d = Acosh(k(h+k))cos(kL)cos(wt)

Luego se debe cumplir que potencial_i=potencial_d

Acosh(k(h+k))cos(wt) = Acosh(k(h+k))cos(kL)cos(wt)
1 = cos(kL)

Las multiples soluciones son 0, pi, pi...

finalmente:  k=(pi)n/L con n =  1, 2, 3, ...


2.4) Relacionando, una condición de borde con el periodo se obtiene:

w^2=g k tanh(kh)

donde sabemos  que w=2pi/T 

(2pi/T)^2=(g k tanh(kh)), reordenando T= (2pi)/(g k tanh(kh))^0.5

en 2.3 se obtuvo la relación "k=(pi)n/L con n =  1, 2, 3, ..." aplicando esta relación en la ecuación anterior se llega a lo pedido.

T= (2pi)/(g ((pi)n/L) tanh(((pi)n/L)h)^0.5
T= 2.pi/{(g.pi.n/L)tanh(pi.n.h/L)}^0.5


2.5) Si h/L tiende a 0 se tiene que tanh(0)= sinh(0)/cosh(0)=0/1=0, pi, 2pi...

tanh(0)= pi n = 0  con n = 1, 2, 3...

como, es igual a 0, podemos decir que "pi.n = pi.n.h/L = 0x0=0"
reemplazando en la equación anterior de 2.4 tanh(0)= pi n h/L

T= 2.pi/{(g.pi.n.h/L^2)tanh(0)}^0.5
T= 2.pi/{(g.pi.n.h/L^2)pi.n}^0.5 
T= 2.pi/{g.pi^2.h.n^2/L^2}^0.5

T= 2L/(n(gh)^0.5) 

Esta ecuación se conoce como la ecuación de Meriam

 Usamos un eta, de stokes que por medio de un armonico de menor amplitud y periodo, simular una onda no lineal de stokes. donde la cresta de la  ola es más "punteaguda" que la parte valle.
 se puede observar una trayectoria de la particula que favorece la subida de la ola que la bajada, ya que esta la empuja hacia la derecha. y cae a la izquierda cuando esta particula desciende.
 
 

```
H=5.0
T=10
L=1.56*(T**2)
h=30
k=2*np.pi/L
x=np.arange(0,4*L+1,1)
w=2*np.pi/T
```

```
x1mem=[]
z1mem=[]
x1=5;
x1i=x1
z1=0
a=H/2
t_total=120  #tiempo de animacion
for t in range(t_total):

  eta= (a)*np.cos(k*x-w*t/2) + k*a**2*np.cosh(k*h)/(4*np.sinh(k*h)**3)*(2+np.cosh(2*k*h))*np.cos(2*(-w*t/2+k*x))
  
  C=-(a)*np.cosh(k*(h+z1))/np.sinh(k*h)*np.sin(k*x1-w*t/2)
  E=(a)*np.sinh(k*(h+z1))/np.sinh(k*h)*np.cos(k*x1-w*t/2)
  C2=w*(a)*np.cosh(k*(h+z1))/np.sinh(k*h)*np.cos(k*x1-w*t/2)
  E2=w*(a)*np.sinh(k*(h+z1))/np.sinh(k*h)*np.sin(k*x1-w*t/2)

  mov = abs(w*k*(2*C*C2 + 2*E*E2))
  xi= -a*np.exp(k*z1)*np.sin(k*x1-w*t/2) + mov
  zi=  a*np.exp(k*z1)*np.cos(k*x1-w*t/2)
  
  x1 += mov
  x1mem.append(x1+xi);z1mem.append(zi)
  
  plt.title('Trayectoria no lineal de particula en olas')
  plt.plot([x1,x1+xi],[z1,z1+zi],'o',x,eta,'lightblue')
  plt.plot(x1mem,z1mem,'gray','--')
  plt.text(60,-4*H,'tiempo = %0.2f'%(t/2,))
  plt.ylim(-h,H)
  plt.xlim(0,L/2)
  plt.title('Trayectoria no lineal de particula en olas')
  plt.xlabel('Largo (m)')
  plt.ylabel('Profundidad (m)')
  plt.pause(0.0001)
  
print('en ',t_total/120,'minutos,')
print("el avance de la particula producto de stokes drift: ", round(x1-x1i,4),'metros')
```
